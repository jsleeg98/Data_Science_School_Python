# 1차원 분포 플롯


```python
import seaborn as sns

iris = sns.load_dataset('iris')
titnanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

```

## 1차원 실수 플롯

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글
mpl.rc('font', family = 'NanumGothic')
mpl.rc('axes', unicode_minus = False)

x = iris.petal_length.values

sns.rugplot(x)
plt.title('Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot')
plt.show()
# 실제 데이터의 위치를 보여준다.
```

```python
sns.kdeplot(x)
plt.title('Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot')
plt.show()
```

```python
sns.distplot(x, kde=True, rug=True)
plt.title('Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot')
plt.show()
```

## 카운트 플롯
카테고리 값별로 데이터가 얼마나 있는지 표시할 수 있다.

```python
sns.countplot(x = 'class', data = titnanic)
plt.title('타이타닉 호의 각 클래스 별, 승객 수 ')
plt.show()
```

```python
sns.countplot(x = 'day', data=tips)
plt.title('요일별 팁을 준 횟수')
plt.show()
```

# 다차원 실수형 데이터
## 2차원 실수형 데이터


```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris)
plt.suptitle('꽃받침의 길이와 넓이의 Joint Plot', y = 1.02)
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde')
plt.suptitle('꽃받침의 길이와 넓이의 Joint Plot과 Kernel Density Plot', y = 1.02)
plt.show()
```
